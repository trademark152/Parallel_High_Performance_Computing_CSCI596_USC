```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline
%pylab inline
```

```python
data_path = ['./pv%d.dat' % i for i in range(1,4)]
pv_data = []
for path in data_path:
    data = np.loadtxt(path)
    pv_data.append(data)
```

```python
matplotlib.rcParams['xtick.direction'] = 'in'
matplotlib.rcParams['ytick.direction'] = 'in'
l1, = plt.plot(pv_data[0][:,0], pv_data[0][:,1])
l2, = plt.plot(pv_data[1][:,0], pv_data[1][:,1])
l3, = plt.plot(pv_data[2][:,0], pv_data[2][:,1])
plt.xlabel('v')
plt.ylabel('P(v)')
xticks = np.linspace(0, 5, 6)
plt.xticks(xticks)
plt.xlim((0,5))
plt.ylim((0,7))
plt.legend(handles=[l1,l2,l3], labels=['10 step','20 step','30 step'])
```
