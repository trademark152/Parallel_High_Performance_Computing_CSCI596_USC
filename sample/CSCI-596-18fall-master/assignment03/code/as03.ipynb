```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
from pylab import *
%matplotlib inline
```

```python
data_path = ['./pv%d.dat' % i for i in range(1,4)]
pv_data = []
for path in data_path:
    data = np.loadtxt(path)
    pv_data.append(data)
```

```python
subplot(1,1,1)
l1, = plot(pv_data[0][:,0], pv_data[0][:,1])
l2, = plot(pv_data[1][:,0], pv_data[1][:,1])
l3, = plot(pv_data[2][:,0], pv_data[2][:,1])
legend(handles=[l1,l2,l3], labels=['10 step','20 step','30 step'])
minorticks_on()
tick_params(which='major',direction='in')
xlim((0,5))
ylim((-0.1,7))
xlabel('v')
ylabel('P(v)')
savefig('./v_Pv.jpg',dpi=200)
```
