```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
t = np.array([1.846639e+01, 1.422450e+01, 7.438687e+00, 6.849241e+00])
threads = np.array([1,2,4,8])
sp = t/t[0]
```

```python
eff = np.divide(sp,threads)
```

```python
plt.plot(threads,eff,'o-')
plt.xlabel('# of threads')
plt.ylabel('Efficiency')
plt.xlim([0,9])
plt.ylim([0,1.2])
plt.plot([0,9],[1,1],'r',linewidth='2.0')
plt.savefig('./eff_thrd.jpg',dpi=400)
```
